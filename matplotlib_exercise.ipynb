## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv("data/hist_data.csv")


plt.hist(df['aデータ'],bins=30,color='orange',label='aデータ',alpha = 0.5)

plt.hist(df['bデータ'],bins=30,color='c',label='bデータ',alpha = 0.5 )

plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')

plt.legend()
plt.show()



```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df =  pd.read_csv("data/scatter_data.csv")

class_0 = df[df["class"] == 0]
class_1 = df[df["class"] == 1]
class_2 = df[df["class"] == 2]

plt.scatter(class_0["length (cm)"], class_0["width (cm)"], label="class 0", c="red")
plt.scatter(class_1["length (cm)"], class_1["width (cm)"], label="class 1", c="blue")
plt.scatter(class_2["length (cm)"], class_2["width (cm)"], label="class 2", c="green")


plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('クラス０～２のlengthとwidthに関する散布図')

plt.legend()
plt.grid()
plt.show()

```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df =  pd.read_csv("data/line_data.csv")
plt.plot(df["期間"],df["件数"],marker = 'o')

plt.xticks(rotation = 30)

plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')

plt.legend()
plt.grid()
plt.show()

```

```python
df =  pd.read_csv("data/line_data.csv")
plt.bar(df["期間"],df["費用"])


plt.xticks(rotation = 30)

plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.title('期間ごとの費用の推移')

plt.legend()
plt.grid()
plt.show()

```
